```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# burasi jupyter noebook icin sadece
%matplotlib inline 
```

```python
kiloList = [15,25,65,45,55,63,75,55,95,75]
```

```python
yasList = [15,18,23,32,43,45,56,58,64,69]
```

```python
numpyYasList = np.array(yasList) # bu verilerle numpy ile biseyler yapabiliriz bu yuzden genelde matplotlib ile yaptigimiz islemleri numpy dizileri uzerinden yapariz
```

```python
numpyKiloList = np.array(kiloList)
```

```python
plt.plot(numpyYasList,numpyKiloList,"g") # g=green,r=red,b=blue vb demek, direk renk ismide yazilabilir
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilonun yasa gore degisim grafigi")
#plt.show() baska bi ide de calisiyorsak bunu kullaniriz
```

```python
plt.plot(numpyYasList,numpyKiloList,"go-") # bu sekilde verilerin kesistigi yerleride gosterebiliriz,*-,x- vs gibi
```

```python
plt.subplot(1,2,1) # 1 sira 2 kolon ve 1.grafik
plt.plot(numpyYasList,numpyKiloList,"rx-")
plt.subplot(1,2,2) # 1 sira 2 kolon ve 2.grafik
plt.plot(numpyKiloList,numpyYasList,"b--")
```

## figure

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.2,0.2,0.9,0.9]) # ilk ikisi konumu diger ikisi tablonunu width v height ayarlari
figureAxes.plot(numpyYasList,numpyKiloList,"y")
figureAxes.set_xlabel("x ekseni")
figureAxes.set_ylabel("y ekseni")
figureAxes.set_title("grafik baslik")
```

## ic ice grafikler

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.2,0.3,0.3,0.3])

eksen1.plot(numpyYasList,numpyKiloList)
eksen1.set_xlabel("x ekseni")
eksen1.set_ylabel("y ekseni")
eksen1.set_title("ana grafik baslik")

eksen2.plot(numpyYasList,numpyKiloList)
eksen2.set_xlabel("x ekseni")
eksen2.set_ylabel("y ekseni")
eksen2.set_title("kucuk grafik baslik")
```

## subplots

```python
(A,B) = plt.subplots(nrows=2,ncols=5)
plt.tight_layout() # tablolar arasi mesafeyi duzgun ayarlicaktir
```

```python
(benimFigure,benimEksenler) = plt.subplots(nrows=1,ncols=2)

for i in benimEksenler:
    i.plot(numpyYasList,numpyKiloList)
    i.set_xlabel("x ekseni")
    
plt.tight_layout() # tablolar arasi mesafeyi duzgun ayarlicaktir
```

## Gorsel Gelistirmeler

```python
numpyDizisi = np.arange(0,10)
numpyDizisi2 =numpyDizisi**2
yeniFigur = plt.figure() # (figsize=(5,4)) diyerek boyutuyla oynayabiliriz, dpi=100 vs diyerekte cozunurlugu arttirabiliriz
yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi,numpyDizisi2,label="Numpy Dizisi**2")
yeniEksen.plot(numpyDizisi,numpyDizisi**3,label="Numpy Dizisi**3")
yeniEksen.legend() # legend ile grafik icinde bilgi verebiliyoruz sol ustte mesela, ama farkli yerede koyabiliriz(loc=1) gibi
```

```python
yeniFigur.savefig("figur.png",dpi=200) #figuru resim olarak kaydetme
```
